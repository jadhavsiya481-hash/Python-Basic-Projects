```python
#Q10

import numpy as np

marks = np.random.randint(30,101,(10,5))
print("Student Marks:")
print(marks)

total_marks = np.sum(marks, axis=1)
avg = np.mean(marks, axis=1)

print("Student Result:")
for i in range(10):
    print(f"Student {i+1}: Total Marks = {total_marks[i]}, Average Marks = {avg[i]:.2f}")

highest = np.argmax(total_marks)
lowest = np.argmin(total_marks)

print(f"Highest Total Marks: Student {highest+1} ({total_marks[highest]})")
print(f"Lowest Total Marks: Student {lowest+1} ({total_marks[lowest]})")
print(f"Class Mean: {np.mean(total_marks):.2f}")
print(f"Class Standard Deviation: {np.std(total_marks):.2f}")
print(f"Marks of top 3 students:")

top3 = np.argsort(total_marks)[-3:]
print(marks[top3])
```
