```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("insurance.csv")
```

```python
df.head(10)
```

```python
print(df.shape)

df.info()

df.describe()
```

```python
df.isnull().sum()
```

```python
numerical = df.select_dtypes(include=['int64', 'float64'])
print(numerical.columns)

categorical = df.select_dtypes(include=['object'])
print(categorical.columns)
```

```python
#5
plt.figure(figsize=(6,4))
sns.histplot(df['age'])
plt.title('Age Distribution')
plt.show()

plt.figure(figsize=(6,4))
sns.histplot(df['bmi'])
plt.title('BMI Distribution')
plt.show()

plt.figure(figsize=(6,4))
sns.histplot(df['children'])
plt.title('childeren Distribution')
plt.show()

plt.figure(figsize=(6,4))
sns.histplot(df['expenses'])
plt.title('expenses Distribution')
plt.show()
```

```python
#6
plt.figure(figsize=(6,4))
sns.countplot(x='sex', data=df)
plt.title('sex count')
plt.show()

plt.figure(figsize=(6,4))
sns.countplot(x='smoker', data=df)
plt.title('smoker count')
plt.show()

plt.figure(figsize=(6,4))
sns.countplot(x='region', data=df)
plt.title('region count')
plt.show()
```

```python
#7
num_df = df.select_dtypes(include=['int64', 'float64'])
plt.figure(figsize=(10,8))
sns.heatmap(
    num_df.corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title('Correlation Heatmap')
plt.show()

```

```python
#8
print(df["expenses"].mean())

print(df["expenses"].max())

print(df["expenses"].min())

df.groupby("smoker")["expenses"].mean()
```

```python
#9
plt.figure(figsize=(6,4))
sns.boxplot(x='smoker', y='expenses', hue='sex', data=df)
plt.title('expenses wrt smoker and sex')
plt.show()
```
