```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("insurance.csv")
print("First 10 rows")
print(df.head(10))
print("Shape")
print(df.shape)
print("Info")
print(df.info())
```

```python
print("Any Missing Values:")
print(df.isnull().sum())
```

```python
print("Statistical Data Summary")
print(df.describe())
```

```python
df[["expenses","bmi","age","children"]].hist(figsize=(10,10))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,3,1)
sns.countplot(x="sex",data=df)

plt.subplot(1,3,2)
sns.countplot(x="smoker",data=df)

plt.subplot(1,3,3)
sns.countplot(x="region",data=df)

plt.tight_layout()
plt.show()
```

```python
num_df = df.select_dtypes(include=["number"])
sns.heatmap(num_df.corr(),annot=True,cmap="coolwarm")
plt.show()
```

```python
print("Columns:")
print(df.columns)

print("Input Features:")
print(df.drop("expenses",axis=1).columns)

print("Target Variable:")
print(df["expenses"].name)
```

```python
df_encoded = pd.get_dummies(df, columns=["sex","smoker","region"])
df_encoded.head()
```

```python
from sklearn.preprocessing import StandardScaler

X = df_encoded.drop("expenses",axis=1)

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled,columns=X.columns)
X_scaled.head()
```

```python
#SUMMARY
# The dataset contains info about the insurance policy holders.
# There are no missing values in the dataset.
# Therefore, the dataset is clean, requires no missing value treatment, and after one hot encoding and feature scaling it is
# ready for building machine learning models such as linear regression.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("apple_products_pricing.csv")
print("First 10 rows")
print(df.head(10))
print("Shape")
print(df.shape)
print("Info")
print(df.info())
```

```python
print("Missing Values")
print(df.isnull().sum())
```

```python
print("Statistical Data Summary")
print(df.describe())
```

```python
df[['Launch_Price_USD','Current_Price_USD']].hist(figsize=(10,8))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,3,1)
sns.countplot(x="Platform",data=df)

plt.subplot(1,3,2)
sns.countplot(x="Product_Category",data=df)

plt.subplot(1,3,3)
sns.countplot(x="Condition",data=df)

plt.tight_layout()
plt.show()
```

```python
num_df = df.select_dtypes(include=["number"])
sns.heatmap(num_df.corr(),annot=True,cmap="coolwarm")
plt.show()
```

```python
print("Columns:")
print(df.columns)

print("Numerical Features:")
print(df.select_dtypes(include="number").columns)

print("Categorical Features:")
print(df.select_dtypes(include="object").columns)
```

```python
df_encoded = pd.get_dummies(df,drop_first=True)
df_encoded.head()
```

```python
from sklearn.preprocessing import StandardScaler

X = df_encoded.drop("Current_Price_USD",axis=1)

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled,columns=X.columns)
X_scaled.head()
```

```python
#SUMMARY
# The dataset contains info about apple products, including their prices and other product details.
# Missing values help identify any incomplete records, as not found in this case.
# Overall, now the data is ready for analysis and predictive modelling.
```
